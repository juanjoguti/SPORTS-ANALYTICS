# Get players data from ongoing season
Import libraries:

```python
import requests
import pandas as pd
from football.constants.seasons import Seasons
```

Request data to the API (should be running at *http://127.0.0.1:8000/docs*):

```python
response = requests.get(f'http://127.0.0.1:8000/v1/football/{Seasons.S2223}/players')
```

Convert from *JSON* to *DataFrame*:

```python
data_season_22_23 = pd.DataFrame.from_dict(response.json())
```
